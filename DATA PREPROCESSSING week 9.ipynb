```python
def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers
```

```python
outliers = find_outliers_IQR(df[["PM2.5", "PM10", "NO", "NO2", "NOx", "CO", "SO2", "O3", "Benzene", "Toluene", "Xylene", "AQI"]])
print("number of outliers:" + str(len(outliers)))
print("max outlier value:" + str(outliers.max()))
print("min outlier value:" + str(outliers.min()))
outliers
```

```python
def impute_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   upper = df[~(df>(q3+1.5*IQR))].max()

   lower = df[~(df<(q1-1.5*IQR))].min()

   df = np.where(df > upper,

       df.mean(),

       np.where(

           df < lower,

           df.mean(),

           df

           )

       )

   return df
```

```python
df[["PM2.5", "PM10", "NO", "NO2", "NOx", "CO", "SO2", "O3", "Benzene", "Toluene", "Xylene", "AQI"]]=impute_outliers_IQR(df[["PM2.5", "PM10", "NO", "NO2", "NOx", "CO", "SO2", "O3", "Benzene", "Toluene", "Xylene", "AQI"]])
```

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
city_label = df.City.value_counts().nlargest(10)

''' barplot '''
plt.figure(figsize=(10, 5))
plt.xticks(rotation=75)
sns.barplot(city_label.index, city_label)
plt.xlabel('City', fontsize=20)
plt.ylabel('Count', fontsize=20);
```

```python
# Create the plot
sns.boxplot(x='City', y='AQI', hue='City', data=df)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))

mask = np.triu(df.corr(method='pearson'))
sns.heatmap(df.corr(method='pearson'),
                 annot=True, fmt='0.1f',
                 mask=mask,
                 robust=True,
                 cmap='pink')
plt.title('Correlation Analysis', fontsize=18);
plt.show()
```

```python
# Group the data by city and calculate the mean PM2.5 value
city_pm25 = df.groupby('City')['PM2.5'].mean()

# Define labels, sizes, and colors for the pie chart
labels = list(city_pm25.index)
sizes = list(city_pm25.values)
colors = ['red', 'orange', 'green']

# Create a pie chart with colors
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add annotations to the chart
for i, label in enumerate(labels):
    angle = (sizes[i]/sum(sizes)) * 360
    ax1.annotate(label, xy=(0, 0), xytext=(1.2*np.cos(np.radians(90-angle/2)), 1.2*np.sin(np.radians(90-angle/2))), fontsize=0, ha='center', va='center', rotation=-(90-angle/2), bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Add a title to the chart
plt.title('PM2.5 values of different cities')

# Display the chart
plt.show()
```
